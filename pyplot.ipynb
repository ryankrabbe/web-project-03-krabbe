### Question 1

### Question 2
Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both `scatter` and `plot` on different axes.  You may use the index of the value as its x value in the plot.

```python
from random import randint
import matplotlib.pyplot as plt

# Generate Random Numbers
random_numbers = [randint(1, 10) for _ in range(20)]
x_values = list(range(len(random_numbers)))

# Subplots for Scatter and Line plot
fig, (plt1, plt2) = plt.subplots(1,2)

# Create Scatter Plot
plt1.scatter(x_values, random_numbers, color='red')
plt1.set_title('Scatter Plot of Random Numbers')
plt1.set_xlabel('Index')
plt1.set_ylabel('Random Number')

#Create Line Plot
plt2.plot(x_values, random_numbers, color='blue')
plt2.set_title('Line Plot of Random Numbers')
plt2.set_xlabel('Index')
plt2.set_ylabel('Random Number')

plt.tight_layout()
plt.show()
```

### Question 3
Do the same question as above, but change the style and color of both the scatter and the plot.

```python
from random import randint
import matplotlib.pyplot as plt

# Generate Random Numbers
random_numbers = [randint(1, 10) for _ in range(20)]
x_values = list(range(len(random_numbers)))

# Subplots for Scatter and Line plot
fig, (plt1, plt2) = plt.subplots(1,2)
fig.patch.set_facecolor('#d1d9e6')

# Create Scatter Plot
plt1.scatter(x_values, random_numbers, color='black', marker='^')
plt1.set_title('Scatter Plot of Random Numbers')
plt1.set_xlabel('Index')
plt1.set_ylabel('Random Number')


#Create Line Plot
plt2.plot(x_values, random_numbers, color='green', marker='s')
plt2.set_title('Line Plot of Random Numbers')
plt2.set_xlabel('Index')
plt2.set_ylabel('Random Number')

plt.tight_layout()
plt.show()
```

### Question 4
Given the data sets below, plot or scatter both on the same set of axes.  You are plotting the time it takes to execute a sorting algorithm on a list of a given size; your x axis should be the size, and the y axis is time.  Assume that `insertion_sort_times[i]` is how long it took to sort a list of length `sizes[i]`.  Add a legend to your plot showing which is insertion sort and which is merge sort.  Use the variable names to determine which data is which. 

```python
sizes = [10, 20, 50, 100, 1000, 5000]
insertion_sort_times = [0.0010252999999999998, 0.0027026999999999954, 0.010147200000000002, 0.0381137, 3.6303399, 91.2180796]
merge_sort_times = [0.00161889999999687, 0.003635600000009731, 0.0090655000000055, 0.020108000000007564, 0.2687474000000094, 1.6147050999999948]

# Create figure and axes
fig, ax = plt.subplots(figsize=(10, 5))
fig.patch.set_facecolor('#d1d9e6')

# Plot the data on the same axes
ax.plot(sizes, insertion_sort_times, color='orange', marker='o', label='Insertion Sort')
ax.plot(sizes, merge_sort_times, color='brown', marker='x', label='Merge Sort')

# Set titles and legend
ax.set_title('Insertion Sort Times vs Merge Sort Times')
ax.set_xlabel('List Size')
ax.set_ylabel('Time (seconds) ')
ax.legend()

# Set y-axis to logarithmic scale
ax.set_yscale('log')

# Display plot
plt.tight_layout()
plt.show()
```
