### Question 2
Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both `scatter` and `plot` on different axes.  You may use the index of the value as its x value in the plot.

```python
from random import randint
import matplotlib.pyplot as plt

# Generate Random Numbers
random_numbers = [randint(1, 10) for _ in range(20)]
x_values = list(range(len(random_numbers)))

# Subplots for Scatter and Line plot
fig, (plt1, plt2) = plt.subplots(1,2)

# Create Scatter Plot
plt1.scatter(x_values, random_numbers, color='red')
plt1.set_title('Scatter Plot of Random Numbers')
plt1.set_xlabel('Index')
plt1.set_ylabel('Random Number')

#Create Line Plot
plt2.plot(x_values, random_numbers, color='blue')
plt2.set_title('Line Plot of Random Numbers')
plt2.set_xlabel('Index')
plt2.set_ylabel('Random Number')

plt.tight_layout()
plt.show()
```

Do the same question as above, but change the syle and color of both the scatter and the plot.

```python
from random import randint
import matplotlib.pyplot as plt

# Generate Random Numbers
random_numbers = [randint(1, 10) for _ in range(20)]
x_values = list(range(len(random_numbers)))

# Subplots for Scatter and Line plot
fig, (plt1, plt2) = plt.subplots(1,2)

# Create Scatter Plot
plt1.scatter(x_values, random_numbers, color='black', marker='^')
plt1.set_title('Scatter Plot of Random Numbers')
plt1.set_xlabel('Index')
plt1.set_ylabel('Random Number')

#Create Line Plot
plt2.plot(x_values, random_numbers, color='green', marker='o')
plt2.set_title('Line Plot of Random Numbers')
plt2.set_xlabel('Index')
plt2.set_ylabel('Random Number')

plt.tight_layout()
plt.show()
```
